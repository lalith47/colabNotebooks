```
def rec(n,memoiz):
  if memoiz[n] is not None:
    return memoiz[n]
  if(n==1 or n==0):
    return n
  
  else:
    i=rec(n-1,memoiz)+rec(n-2,memoiz)
    memoiz.append(i)
    return i
```

```
def fib_memo(n):
  memoiz=[None]*(n+1)
  return rec(n,memoiz)
```

```
fib_memo(100)
```

```
def count_ways(n,c):
  l=[0]* (len(c)+1)
  l[0]=1
  for i in range(1,n+1):
    for j in c:
      l[i]+=l[i-j]
```

```
count_ways(4,[1,2,3])
```

```
def ways_dp(n,c):

  dp=[None]*(n+1)
  def no_of_ways(n,dp):
    sum=0
    if(n<0):
      return 0
    if(n==1 or n==0):
      return 1
    if(dp[n]!=None):
      return dp[n]
    for i in c:
      sum+=no_of_ways(n-i,dp)
    dp[n]=sum
    return dp[n]
  return no_of_ways(n,dp)
```

```
ways_dp(4,[1,2,3])
```

```
class TreeNode:
  def __init__(self,x):
    self.val=x
    self.right=None
    self.left=None
  def __str__(self):
    return "TreeNode{val: "+str(self.val)+", left: "+str(self.left)+", right: "+str(self.right)+"}"
```

```
root=TreeNode(1)
root.left=TreeNode(2)
root.right=TreeNode(3)
root.left.left=TreeNode(4)
root.left.right=TreeNode(5)
root.right.right=TreeNode(6)
```

```
print(root)
```

```
def shread_tree(tree_node,lst):
  if(tree_node!= None):
    lst.append(tree_node.val)
    if(tree_node.left!=None):
      shread_tree(tree_node.left,lst)
    if(tree_node.right!=None):
      shread_tree(tree_node.right,lst)

```

```
def shread_tree_new(tree_node):
  lst=[]
  if(tree_node!= None):
    lst.append(tree_node.val)
    if(tree_node.left!=None):
      lst+= shread_tree_new(tree_node.left)
    if(tree_node.right!=None):
      lst+=shread_tree_new(tree_node.right)
    return lst
```
